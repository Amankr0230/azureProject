```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window

import os 
import sys

project_pth = os.path.join(os.getcwd(),'..','..')

sys.path.append(project_pth)

from utils.transformations import reusable

```

```python
df = spark.read.format("parquet")\
    .load("abfss://bronze@storageazureproject007.dfs.core.windows.net/DimUser")
```

```python
display(df)
```

**Autoloder**

```python
df_user = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format","parquet")\
            .option("cloudFiles.schemaLocation","abfss://silver@storageazureproject007.dfs.core.windows.net/DimUser/chekpoint")\
            .option("schemaEvolutionMode","addNewColumns")\
            .load("abfss://bronze@storageazureproject007.dfs.core.windows.net/DimUser")
```

```python
display(df_user)
```

```python
df_user = df_user.withColumn("user_name",upper(col("user_name")))
display(df_user)
```

```python
df_user_obj = reusable()

df_user = df_user_obj.dropColumns(df_user,['_rescued_data'])
df_user = df_user.dropDuplicates(['user_id'])
display(df_user)    
```

```python
df_user.writeStream.format("delta")\
            .outputMode("append")\
            .option("checkpointLocation","abfss://silver@storageazureproject007.dfs.core.windows.net/DimUser/chekpoint")\
            .trigger(once=True)\
            .option("path","abfss://silver@storageazureproject007.dfs.core.windows.net/DimUser/data")\
            .toTable("spotify_cata.silver.DimUser")
            
```

## **DimTrack**

```python
df_track = spark.readStream.format("cloudFiles")\
            .option("cloudFiles.format","parquet")\
            .option("cloudFiles.schemaLocation","abfss://silver@storageazureproject007.dfs.core.windows.net/DimTrack/chekpoint")\
            .option("schemaEvolutionMode","addNewColumns")\
            .load("abfss://bronze@storageazureproject007.dfs.core.windows.net/DimTrack")
```

```python
df_track = df_track.withColumn("durationFlag",when(col('duration_sec')<150,"low")\
                                            .when(col('duration_sec')<300,"medium")\
                                            .otherwise("high"))

df_track = df_track.withColumn("track_name",regexp_replace(col('track_name'),'-',' '))

df_track = reusable().dropColumns(df_track,['_rescued_data'])

display(df_track)
```

```python
df_track.writeStream.format("delta")\
            .outputMode("append")\
            .option("checkpointLocation","abfss://silver@storageazureproject007.dfs.core.windows.net/DimTrack/chekpoint")\
            .trigger(once=True)\
            .option("path","abfss://silver@storageazureproject007.dfs.core.windows.net/DimTrack/data")\
            .toTable("spotify_cata.silver.DimTrack")
```
